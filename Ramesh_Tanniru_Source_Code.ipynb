```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Excel file
df = pd.read_excel("/content/1730285881-Airbnb_Open_Data.xlsx")

# Display first few rows
display(df.head())
```

```python
# Count neighbourhood groups
neigh_group_counts = df['neighbourhood group'].value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
neigh_group_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
plt.title("Distribution of Listings by Neighbourhood Group")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['neighbourhood'].value_counts().nlargest(10).plot(kind='bar', color="skyblue")
plt.title("Top 10 Neighbourhoods by Number of Listings")
plt.xlabel("Neighbourhood")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df, x='host_identity_verified', palette="muted")
plt.title("Host Identity Verification Status")
plt.xlabel("Verification Status")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x="long", y="lat", hue="neighbourhood group", alpha=0.5, palette="Set1")
plt.title("Geographical Distribution of Listings")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df.groupby("neighbourhood group")["price"].mean().plot(kind="bar", color="orange")
plt.title("Average Price by Neighbourhood Group")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['host name'].value_counts().nlargest(10).plot(kind='bar', color="teal")
plt.title("Top 10 Hosts by Number of Listings")
plt.xlabel("Host Name")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['room type'].value_counts().plot(kind='bar', color="coral")
plt.title("Room Type Distribution")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="neighbourhood group", y="price", palette="Set3")
plt.ylim(0, 1000)  # Limit to avoid extreme outliers
plt.title("Price Distribution by Neighbourhood Group")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['availability 365'], bins=30, color="purple", kde=False)
plt.title("Distribution of Availability (Days per Year)")
plt.xlabel("Days Available")
plt.ylabel("Number of Listings")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="long", y="lat", hue="price", size="price",
                palette="viridis", alpha=0.6, legend=False)
plt.title("Listings Geographical Distribution Colored by Price")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="host_identity_verified", hue="room type", palette="Set2")
plt.title("Room Types by Host Verification Status")
plt.xlabel("Host Identity Verified")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df.groupby("neighbourhood")["price"].mean().nlargest(15).plot(kind="bar", color="navy")
plt.title("Top 15 Neighbourhoods by Average Price")
plt.ylabel("Average Price")
plt.show()
```
